## **3. Exploratory Data Analysis (EDA)**

- Distribution Analysis
- Correlation Analysis
- Feature Relationships
- Outlier Detection

### **Distribution Analysis**

- Checked how values are spread (evenly or skewed) to see if data needs adjustment. 
- Most features are concentrated in typical ranges. 
- sqft_living shows slight right skew, while other features such as bedrooms and bathrooms follow expected distributions.

```python
# zipcode excluded
features = ["bedrooms", "bathrooms", "sqft_living", "yr_built"]

# plot distributions of selected features
df[features].hist(figsize=(6, 4))
plt.suptitle("Distribution of Selected Features")
plt.tight_layout()
plt.show()
```

### **Correlation Analysis**

- Checking correlation helps understand how features relate to the target (price).
- sqft_living shows the strongest correlation with price, while other features have weaker relationships.

```python
corr = df.corr(numeric_only=True)

plt.figure(figsize=(11, 7))
sns.heatmap(corr, annot=True, cmap='crest')
plt.title("Correlation Matrix")
plt.tight_layout()
plt.show()
```

### **Feature Relationships**

- Based on the correlation analysis above, sqft_living shows a strong relationship with price.
- Visualized their relationship using a scatter plot.
- Clear positive relationship in between living areas and price, supporting the correlation analysis.
- Prices can be different even for homes with similar living areas, meaning other factors also affect price.

```python
sns.scatterplot(x=df["sqft_living"], y=df["price"])
plt.title("Relationship Between Living Area and Price")
plt.show()
```

### **Outlier Detection**

- Performed on key features to find unusual values.
- Values are generally within a normal range, with no significant outliers.
- Some higher values are present in price, but they appear to be valid.
- Key features such as sqft_living and bedrooms were selected based on their importance and relationship with price.

```python
plt.figure(figsize=(6, 4))
sns.boxplot(data=df[["price", "sqft_living", "bedrooms"]])
plt.show()
```
